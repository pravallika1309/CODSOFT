```python
x = credit.drop(columns = 'Class',axis=1)
y = credit['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=13)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
x_train_pred = LR.predict(x_train)
```

```python
training_data_acuracy=accuracy_score(x_train_pred,y_train)*100
print(f"Training Data Accuracy: {training_data_acuracy}%")
```

```python
x_test_pred=LR.predict(x_test)
test_data_accuracy=accuracy_score(x_test_pred,y_test)*100
print(f"Test Data Accuracy: {test_data_accuracy}%")
```
